```python
import cv2

img_circles = cv2.imread('Circles_Noise.bmp', cv2.IMREAD_GRAYSCALE)
img_binary = cv2.imread('Binary_Noise.bmp', cv2.IMREAD_GRAYSCALE) 
```

```python
import matplotlib.pyplot as plt
def plot(original_image, binarized_image):
    fig, axs = plt.subplots(1, 2, figsize=(13, 13))
    axs[0].imshow(original_image, cmap='gray')
    axs[0].set_title('Original')
    axs[1].imshow(binarized_image, cmap='gray')
    axs[1].set_title('Agregado de cores')
    plt.show()
```

```python
_, axs = plt.subplots(1, 2, figsize=(13, 13))

axs[0].imshow(img_binary, cmap='gray')
axs[1].imshow(img_circles, cmap='gray') 
```

as imagem aparentam ter um rudo no padro sal e pimenta e com isso foi escolhido o filtro mediana para fazer a filtragem passa baixa na imagem. J que  uma boa opo de filtro para esse tipo de caso.

```python
img_1 = cv2.medianBlur(img_circles, 5)
```

```python
img_2 = cv2.medianBlur(img_binary, 5) 
# img_2 = cv2.GaussianBlur(img_binary, (5, 5), 0)  
```

```python
_, axs = plt.subplots(1, 2, figsize=(13, 13))

axs[0].imshow(img_1, cmap='gray')
axs[1].imshow(img_2, cmap='gray') 
```
