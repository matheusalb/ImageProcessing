```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img_book_1 = cv2.imread('Book_1.bmp', cv2.IMREAD_GRAYSCALE)
img_book_2 = cv2.imread('Book_2.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
# Imagens j√° binarizadas
np.unique(img_book_1), np.unique(img_book_2)
```

```python
# Inverte a cor dos pixels
img_book_1 = cv2.bitwise_not(img_book_1)
img_book_2 = cv2.bitwise_not(img_book_2)
```

```python
plt.imshow(img_book_1, cmap='gray')
```

```python
def exist_A(img, A_format):
    erode_img = cv2.erode(img, A_format, iterations=1)
    if np.any(erode_img == 255):
        print("Existe A")
    else:
        print("N√£o existe A")
    
    plt.imshow(erode_img, cmap='gray')
    return erode_img
```

```python
img_A = cv2.imread('A.jpeg', cv2.IMREAD_GRAYSCALE)

img_A = (img_A < 127).astype('uint8') # Normaliza a imagem entre 0 e 1 e inverte cor
```

```python
np.unique(img_A)
```

```python
plt.imshow(img_A, cmap='gray')
```

```python
np.unique(img_A)
```

```python
erode_img_1 = exist_A(img_book_1, img_A)
```

```python
np.unique(erode_img_1, return_counts=True)
```

Como extra, com essa abordagem √© poss√≠vel contar tamb√©m a quantidade de 'A' presente na imagem. No caso temos 10 valores de pixel 255 para essa imagem, ou seja, 10 vezes que a elemento elemento estruturante teve encontro de padr√£o na imagem, dessa forma, a imagem possui 10 'A's, valor que foi confirmado visualmente.

```python
erode_img_2 = exist_A(img_book_2, img_A)
```
